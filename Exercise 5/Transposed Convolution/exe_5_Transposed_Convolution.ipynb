```python
from numpy import asarray
from keras.models import Sequential
from keras.layers import Conv2DTranspose
```

```python
# define input data
X = asarray([[1, 2],
             [3, 4]])

# show input data for context
print(X)

# reshape input data into one sample a sample with a channel
X = X.reshape((1, 2, 2, 1))

# define model
model = Sequential()
model.add(Conv2DTranspose(1, (2,2), strides=(2,2), padding='same', input_shape=(2, 2, 1)))
model.summary()

#define weights that mimic nearest neighbour sampling
weights = [asarray([[[[1, 1],
                      [1, 1]]]]), asarray([0])]
weights[0] = weights[0].reshape(2,2,1,1)

model.set_weights(weights)
yhat = model.predict(X)

# reshape output to remove channel to make printing easier
yhat = yhat.reshape((4, 4))
print(yhat)
```

```python
import cv2
from matplotlib import pyplot as plt
from numpy import asarray
from keras.models import Sequential
from keras.layers import Conv2DTranspose

#
# define model
model = Sequential()
model.add(Conv2DTranspose(1, (4,4), strides=(2,2), padding='same', input_shape=(2, 2, 1)))
model.summary()

#define weights that mimic nearest neighbour sampling
weights = [asarray([[[
    [0.0625, 0.1875, 0.1875, 0.0625],
    [0.1875, 0.5625, 0.5625, 0.1875],
    [0.1875, 0.5625, 0.5625, 0.1875],
    [0.0625, 0.1875, 0.1875, 0.0625]
    ]]]), asarray([0])]
weights[0] = weights[0].reshape(4,4,1,1)
model.set_weights(weights)
#

img_bgr = cv2.imread('/content/example.png',cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis("off")
plt.show()

img_r = img_bgr[:,:,2]
img_g = img_bgr[:,:,1]
img_b = img_bgr[:,:,0]
img_r = img_r.reshape((1, 32, 32, 1))
img_g = img_g.reshape((1, 32, 32, 1))
img_b = img_b.reshape((1, 32, 32, 1))

up_r = model.predict(img_r)
up_g = model.predict(img_g)
up_b = model.predict(img_b)

up_r = up_r.reshape((64, 64))
up_g = up_g.reshape((64, 64))
up_b = up_b.reshape((64, 64))

img_up = cv2.merge((up_r, up_g, up_b)).astype('uint8')

plt.imshow(img_up)
plt.axis("off")
plt.show()
```
